## Bubble Sort Algorithm
![alternatvie text](https://files.realpython.com/media/python-sorting-algorithms-bubble-sort.216ab9a52018.png)

```python
# Bubble Sort implementation
def bubble_sort(array):
    n = len(array)

    for i in range(n):
        for j in range(n - i - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
    return array

array = [8, 2, 6, 4, 5,7,9,12,54,23,90,23,67,43]
print(bubble_sort(array))
```

## Insertion Sort
![alternative_text](https://files.realpython.com/media/python-sorting-algorithms-insertion-sort.a102f819b3d7.png)

```python
#insertion sort
def insertion_sort(array):
    for i in range(1, len(array)):
        key_item = array[i]
        j = i - 1
        while j >= 0 and array[j] > key_item:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = key_item

    return array

array = [8, 2, 6, 4, 5,7,9,12,54,23,90,23,67,43]
print(bubble_sort(array))
```

## Merge Sort Algorithm
![merge_sort](https://robocrop.realpython.net/?url=https%3A//files.realpython.com/media/python-sorting-algorithms-merge-sort.d6b5c7dec9ef.png&w=851&sig=54c49f27a8be08b67009b3a2daf7c9e93e7e7593)

```python
#implementation of merge funcation
def merge(left, right):

    if len(left) == 0: #first array is empty
        return right
    if len(right) == 0: #second arrary is empty
        return left

    result = []
    index_left = index_right = 0
    while len(result) < len(left) + len(right):
        if left[index_left] <= right[index_right]: #if element of left-array is lower
            result.append(left[index_left])
            index_left += 1
        else:
            result.append(right[index_right]) #if element of right-array is lower
            index_right += 1

        if index_right == len(right):   #if reached to the end of right array
            result += left[index_left:]
            break

        if index_left == len(left):  #if reached to the end of left array
            result += right[index_right:]
            break

    return result
```

```python
#Implementation of merge_sort
def merge_sort(array):
    if len(array) < 2:
        return array

    midpoint = len(array) // 2
    return merge(
        left=merge_sort(array[:midpoint]),
        right=merge_sort(array[midpoint:]))


array = [8, 2, 6, 4, 5,7,9,12,54,23,90,23,67,43]
print(merge_sort(array))
```

## Quick Sorty Implementation
![Quick_Sort](https://files.realpython.com/media/Python_Sorting_Algorithms_-_Quick_Sort.ee6dbe24f0d3.jpeg)

```python
#implementation of quick sort
from random import randint

def quicksort(array):
    # If the input array contains fewer than two elements,
    # then return it as the result of the function
    if len(array) < 2:
        return array

    low, same, high = [], [], []

    # Select your `pivot` element randomly
    pivot = array[randint(0, len(array) - 1)]

    for item in array:
        if item < pivot:
            low.append(item)
        elif item == pivot:
            same.append(item)
        elif item > pivot:
            high.append(item)

    return quicksort(low) + same + quicksort(high)

array = [8, 2, 6, 4, 5,7,9,12,54,23,90,23,67,43]
print(quicksort(array))
```
